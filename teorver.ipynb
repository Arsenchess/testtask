3. Задача по теории вероятности:

Условие: В коробке 7 красных, 5 зеленых и 8 синих шаров. Наугад выбирается один шар. Какова вероятность того, что выбранный шар окажется либо красным, либо зеленым?

Ожидаемый результат: Описание решения и ответ

Находим общее количество шаров в коробке:

Общее количество шаров = (количество красных) + (количество зеленых) + (количество синих)

Общее количество шаров = 7+5+8=20 шаров.

Находим количество благоприятных исходов, красный или зеленый

Количество благоприятных исходов = (количество красных шаров) + (количество зеленых шаров)

Количество благоприятных исходов = 7+5=12 шаров.

формула вероятности

Р(событие) = колво благоприятных исходов/общее колво возможнных исходов

у нас это либо красный или зеленый

Р = колво красных или зеленых/сумма всех шаров

Р = 12/20

Упрощаем дробь

12/20 = 3x4/5x4 = 3/5

переводим в десятичную дробь и потом в проценты

3/5 = 0.6

0.6 x 100% = 60 %

Ответ:

Вероятность того, что выбранный шар окажется либо красным, либо зеленым будет 3/5 или 0.6 (60%)
