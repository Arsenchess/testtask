1. Задача по написанию тест-кейсов:

Условие: В веб-приложении есть выпадающий список с выбором стран и кодом телефона. Необходимо проверить корректность работы списка.

Ожидаемый результат: Описание позитивных и негативных тест-кейсов.

Контекст для оформления баг-репортов
Название тест-кейса или описание проблемы: Кратко и точно.

Шаги для воспроизведения: Подробное описание действий, которые привели к ошибке.

Ожидаемый результат: Что должно было произойти.

Фактический результат: Что произошло на самом деле.

Серьезность/Приоритет: Насколько критичен баг.

Скриншоты/Видео: Визуальное подтверждение проблемы.

Окружение, в котором был обнаружен баг:

Браузер и его версия: (например, Chrome 125.0.6422.112, Firefox 126.0)

Версия билда/сборки приложения: (например, build 2.5.1-beta)

Версия OS: (например, Windows 11 Pro 23H2, macOS Sonoma 14.5, Android 14, iOS 17.5)

Воспроизводимость: Как часто баг воспроизводится (например, "Всегда (3 из 3 попыток)", "Иногда (1 из 5 попыток)", "Не удалось воспроизвести повторно").

Дополнительные комментарии: Любая другая информация, которая может помочь разработчикам.

Позитивные тест-кейсы (Ожидаемый результат)
Отображение списка по умолчанию:

WTR:

1: Открыть страницу/секцию с выпадающим списком.

Ожидаемый результат: По умолчанию в поле отображается страна/код, заданные по умолчанию (если есть), либо плейсхолдер (например, "Выберите страну"). Список не раскрыт.

Раскрытие списка:

WTR:

1: Кликнуть на поле выпадающего списка.

Ожидаемый результат: Список стран с соответствующими телефонными кодами корректно раскрывается. Все элементы списка видны и читаемы.

Выбор элемента из списка:

WTR:

1: Раскрыть список.

2: Кликнуть на одну из стран в списке.

Ожидаемый результат: Выбранная страна и ее телефонный код корректно отображаются в поле. Список сворачивается. Соответствующее значение (например, country_id или phone_code) передается в систему.

Корректность данных в списке:

WTR:

1: Раскрыть список.

2: Проверить несколько случайных стран.

Ожидаемый результат: Названия стран указаны без ошибок. Телефонные коды соответствуют выбранным странам. Формат кодов корректен (например, "+7", "+380").

Поиск/фильтрация в списке (если есть такая функциональность):

WTR:

1: Раскрыть список.

2: Начать вводить название страны или код в поле поиска (если оно есть).

Ожидаемый результат: Список динамически фильтруется, отображая только те страны, которые соответствуют введенному тексту.

Навигация по списку с помощью клавиатуры:

WTR:

1: Сделать поле выпадающего списка активным (например, кликнув на него или используя Tab).

2: Использовать клавиши "вверх/вниз" для перемещения по элементам списка.

3: Нажать Enter или Space для выбора элемента.

Ожидаемый результат: Навигация работает корректно, выбранный элемент отображается в поле.

Закрытие списка без выбора:

WTR:

1: Раскрыть список.

2: Кликнуть вне области списка.

Ожидаемый результат: Список сворачивается, значение в поле не изменяется (остается предыдущее или значение по умолчанию).

Проверка наличия всех ожидаемых стран:

WTR:

1: Раскрыть список.

2: Проверить наличие ключевых/ожидаемых стран.

Ожидаемый результат: Все необходимые страны присутствуют в списке.

Проверка уникальности элементов:

WTR:

1: Раскрыть список.

2: Просмотреть список на наличие дублирующихся стран или кодов.

Ожидаемый результат: Каждая страна и ее код представлены в списке уникально.

Работа с длинными названиями стран:

WTR:

1: Найти в списке страну с очень длинным названием.

Ожидаемый результат: Название отображается корректно, не обрезается и не нарушает верстку списка или поля.

Негативные тест-кейсы (Актуальный результат, указывающее на баг)
Пустой список:

WTR:

1: Кликнуть на поле выпадающего списка.

Актуальный результат (баг): Список не раскрывается или раскрывается пустым, хотя страны должны быть.

Некорректные данные:

WTR:

1: Раскрыть список.

2: Проверить элементы.

Актуальный результат (баг): Обнаружены ошибки в названиях стран, неправильные телефонные коды, несоответствие страны и кода, наличие спецсимволов там, где их быть не должно.

Невозможность выбрать элемент:

WTR:

1: Раскрыть список.

2: Попытаться кликнуть на одну из стран.

Актуальный результат (баг): Клик не приводит к выбору страны, значение в поле не изменяется.

Неправильное отображение выбранного элемента:

WTR:

1: Выбрать страну из списка.

Актуальный результат (баг): В поле отображается другая страна, неверный код, или значение вообще не отображается.

Проблемы с версткой при раскрытии списка:

WTR:

1: Раскрыть список.

Актуальный результат (баг): Список отображается некорректно: выходит за пределы экрана, перекрывает другие важные элементы, имеет неправильную ширину/высоту, элементы накладываются друг на друга.

Не работает поиск/фильтрация (если есть):

WTR:

1: Ввести текст в поле поиска.

Актуальный результат (баг): Список не фильтруется, или фильтруется некорректно (например, пропадают все элементы или остаются неподходящие).

Не работает навигация с клавиатуры:

WTR:

1: Попытаться использовать клавиши "вверх/вниз" и Enter/Space.

Актуальный результат (баг): Навигация не работает, выбрать элемент с клавиатуры невозможно.

Список не закрывается:

WTR:

1: Раскрыть список.

2: Попытаться закрыть его кликом вне списка или выбором элемента.

Актуальный результат (баг): Список остается открытым.

Отсутствие обязательных стран:

WTR:

1: Раскрыть список.

Актуальный результат (баг): В списке отсутствуют страны, которые должны быть по требованиям.

Дублирование элементов:

WTR:

1: Раскрыть список.

Актуальный результат (баг): Некоторые страны или коды телефонов повторяются в списке.

Проблемы при взаимодействии с другими элементами страницы:

WTR:

1: Раскрыть список.

2: Попытаться взаимодействовать с другими полями ввода или кнопками на странице.

Актуальный результат (баг): Раскрытый список мешает взаимодействию, или взаимодействие с другими элементами приводит к неожиданному поведению списка (например, самопроизвольному закрытию или выбору элемента).

Некорректная работа при изменении масштаба страницы:

WTR:

1: Изменить масштаб страницы в браузере (Ctrl + "+", Ctrl + "-").

2: Попробовать взаимодействовать со списком.

Актуальный результат (баг): Элементы списка отображаются некорректно, список "уезжает", невозможно выбрать элемент.

Некорректная работа на разных разрешениях экрана/устройствах:

WTR:

1: Протестировать на различных разрешениях экрана (например, эмулируя мобильные устройства в браузере).

Актуальный результат (баг): Список не адаптируется под размер экрана, элементы не видны, пользоваться неудобно.

Отправка формы с невыбранным значением (если поле обязательное):

WTR:

1: Оставить поле страны пустым (если это возможно и поле обязательное).

2: Попытаться отправить форму.

Актуальный результат (баг): Форма отправляется без указания страны, либо сообщение об ошибке не появляется/некорректно.

Ввод невалидных данных в поле поиска (если есть):

WTR:

1: Ввести в поле поиска спецсимволы, очень длинную строку, SQL-запрос (для проверки безопасности например).

Актуальный результат (баг): Приложение "падает", отображается ошибка, происходит несанкционированный доступ к данным, или поиск отрабатывает некорректно.
